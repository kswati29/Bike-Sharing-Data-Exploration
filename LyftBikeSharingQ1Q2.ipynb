```python
#Question 1 and 2 are both summary statistics questions to develop the foundation for questions 3-5.
#Question 1 determines that routes within San Francisco city limits are the most frequent. 
#Therefore, we'll be focusing most of our efforts on analyzing the ride data within San Francisco.
#Question 2 shows a distribution of contribution based on gender, age, and user type. 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

## Question 1

```python
#Determine which city has the highest volume and duration of trips. This city will have the highest revenue.
#(Bar graphs)
```

```python
#Add a new row that contains both the start and end city
col_concat = pd.concat([Question_1_2['start_city'] + " / " + Question_1_2['end_city']], axis = 1)
col_concat
```

```python
Start_End = col_concat
```

```python
Question_1_2['start_end_city'] = Start_End
```

```python
Question_1_2
```

```python
#We're wanting to see clearly which route has the highest frequency of riders.
Route_count = Question_1_2.groupby('start_end_city').start_end_city.agg(np.count_nonzero)
```

```python
print(Route_count)
```

```python
#We are counting how many times each of the routes are taken over the 5 month period.
import seaborn as sns
count, ax = plt.subplots()

ax = sns.countplot(x = 'start_end_city', data = Question_1_2, palette = 'Set1')

ax.set_title('Frequency of Route Usage', fontsize='x-large', fontweight='bold')
ax.set_xlabel('Route Name', fontsize='large', fontweight='bold')
ax.set_ylabel('Total Times Taken', fontsize='large', fontweight='bold')
ax.set_xticklabels(
    ax.get_xticklabels(), 
    rotation=35, 
    horizontalalignment='right',
    fontweight='light')

plt.show()
```

```python
#Highest average duration per route. 
Route_Duration = Question_1_2.groupby('start_end_city').trip_duration_sec.agg(np.mean)
```

```python
Route_Duration
```

```python
bar, ax = plt.subplots()
ax = sns.barplot(x = 'start_end_city', y = 'trip_duration_sec', data = Question_1_2, palette = 'Set1')
ax.set_title('Average Duration Per Route', fontsize='x-large', fontweight='bold')
ax.set_xlabel('Route Name', fontsize='large', fontweight='bold')
ax.set_ylabel('Average Duration (in seconds)', fontsize='large', fontweight='bold')
ax.set_xticklabels(
    ax.get_xticklabels(), 
    rotation=35, 
    horizontalalignment='right',
    fontweight='light')

plt.show()
```

```python
# We are calculating the distribution of rides over the months to determine if there are any significant outliers to analyze.
Month_count = Question_1_2.groupby('month').month.agg(np.count_nonzero)
```

```python
Month_count
```

```python
count, ax = plt.subplots()

ax = sns.countplot(x = 'month', data = Question_1_2, palette = 'Set1')

ax.set_title('Frequency Per Month', fontsize='x-large', fontweight='bold')
ax.set_xlabel('Month', fontsize='large', fontweight='bold')
ax.set_ylabel('Total Amount of Rides', fontsize='large', fontweight='bold')
ax.set_xticklabels(
    ax.get_xticklabels(),
    fontweight='light')

plt.show()
```

## Import Lyft website data into separate DataFrames

```python
lyft_jan_2019 = pd.read_csv('201901-fordgobike-tripdata.csv')

lyft_feb_2019 = pd.read_csv('201902-fordgobike-tripdata.csv')

lyft_mar_2019 = pd.read_csv('201903-fordgobike-tripdata.csv')

lyft_apr_2019 = pd.read_csv('201904-fordgobike-tripdata.csv')

lyft_may_2019 = pd.read_csv('201905-baywheels-tripdata.csv')
```

## Combine DataFrames and display row count

```python
lyft_combined_2019 = pd.concat([lyft_jan_2019, lyft_feb_2019, lyft_mar_2019, lyft_apr_2019, lyft_may_2019], ignore_index = True)
lyft_combined_2019.shape[0]
```

## Display null counts of each column

```python
lyft_num_rows = lyft_combined_2019.shape[0]
lyft_num_missing = num_rows - lyft_combined_2019.count()
print(num_missing)
```

## Create DataFrames for both start and end stations and only keep station info. Then, remove rows with any NaNs

```python
start_stations = lyft_combined_2019[ ['start_station_id', 'start_station_latitude', 'start_station_longitude'] ]
start_stations_filtered = start_stations.dropna(how = "any")

end_stations = lyft_combined_2019[ ['end_station_id', 'end_station_latitude', 'end_station_longitude'] ]
end_stations_filtered = end_stations.dropna(how = "any")
```

## Drop any duplicate rows and rename columns for concatenation

```python
all_start_stations = start_stations_filtered.drop_duplicates()
all_start_stations.columns = ['station_id', 'station_latitude', 'station_longitude']

all_end_stations = end_stations_filtered.drop_duplicates()
all_end_stations.columns = ['station_id', 'station_latitude', 'station_longitude']
```

## Combine station DataFrames, drop duplicates, and see if there are any stations appearing more than once

```python
combine_stations = pd.concat([all_start_stations, all_end_stations], ignore_index = True)
unique_stations = combine_stations.drop_duplicates()

group_count = unique_stations.groupby('station_id', sort=True).filter(lambda x: x['station_id'].count() > 1)
group_count.sort_values(by=['station_id'])
```

## Remove reoccuring station records and check for any more

```python
unique_stations_dedupe = unique_stations.drop([362, 332, 326, 364, 324])

unique_stations_dedupe.groupby('station_id', sort=True).filter(lambda x: x['station_id'].count() > 1)
```

## Merge Lyft website station DataFrame onto Kaggle DataFrame to get latitude and longitude of each start and end station

```python
get_start_coordinates = NaN_filtered_combined_2019.merge(unique_stations_dedupe, left_on = 'start_station_id' , right_on = 'station_id', how='left')
get_end_coordinates = get_start_coordinates.merge(unique_stations_dedupe, left_on = 'end_station_id' , right_on = 'station_id', how='left')
get_end_coordinates
```

## Drop unneeded columns and rename some

```python
rides_with_coordinates = get_end_coordinates.drop(['station_id_x', 'station_id_y'], axis = 1)

rename_dict = {'station_latitude_x': 'start_station_latitude', 'station_longitude_x': 'start_station_longitude',
               'station_latitude_y': 'end_station_latitude', 'station_longitude_y': 'end_station_longitude'}

rides_with_coordinates.columns = [rename_dict.get(x,x) for x in rides_with_coordinates.columns]
rides_with_coordinates
```

## Create function to determine city where a ride begins

```python
def determine_start_city(df):
    if df['start_station_longitude'] < -122.354860:
        val = 'San Francisco'
    elif df['start_station_latitude'] < 37.4258460:
        val = 'San Jose'
    elif df['start_station_latitude'] > 37.7492920 and df['start_station_longitude'] > -122.3390910:
        val = 'Oakland/Berkeley'
    else:
        val = 'Unknown'
    return val
```

## Add new start_city column and apply function

```python
rides_with_coordinates['start_city'] = rides_with_coordinates.apply(determine_start_city, axis=1)
rides_with_coordinates
```

## Create function to determine city where a ride ends

```python
def determine_end_city(df):
    if df['end_station_longitude'] < -122.354860:
        val = 'San Francisco'
    elif df['end_station_latitude'] < 37.4258460:
        val = 'San Jose'
    elif df['end_station_latitude'] > 37.7492920 and df['end_station_longitude'] > -122.3390910:
        val = 'Oakland/Berkeley'
    else:
        val = 'Unknown'
    return val
```

## Add new end_city column and apply function

```python
rides_with_coordinates['end_city'] = rides_with_coordinates.apply(determine_end_city, axis=1)
rides_with_coordinates
```

## Remove outlier rows based on 2 criteria:
## 1. Rides more than 6 hours long: could be an incorrect docking or stolen bike
## 2. Rides less than 2 minutes (where station in and out is the same): based on assumption that ride is not working properly and user returns it

```python
# 1. drop rows where ride is greater than 6 hours
drop_long_rides = rides_with_coordinates.drop(rides_with_coordinates[(rides_with_coordinates['trip_duration_sec']>21600)].index)
drop_long_rides
```

```python
# 2. drop rows where ride is less than 2 minutes and station in and out is same
final_total_rides_cleaned = drop_long_rides.drop(drop_long_rides[(drop_long_rides['trip_duration_sec']<=120) & (drop_long_rides['start_station_id'] == drop_long_rides['end_station_id'])].index)
final_total_rides_cleaned
```

## Save cleaned DataFrame as a CSV called: final_total_rides_cleaned.csv

```python
final_total_rides_cleaned.to_csv('final_cleaned_ride_data.csv')
```

# Question 5: Data Model/Multiple Regression

## Start by subsetting to just San Francisco rides

```python
san_francisco_rides = final_total_rides_cleaned[(final_total_rides_cleaned["start_city"] == "San Francisco") & (final_total_rides_cleaned["end_city"] == "San Francisco")]
san_francisco_rides
```

## Next, add a new column called "age" to determine user's age in 2019

```python
san_francisco_rides['age'] = 2019 - san_francisco_rides['member_birth_year']
san_francisco_rides
```

## Create a function to determine a member's age range:

## Possible categories:
## 1. Under 18
## 2. 18-35
## 3. 36-50
## 4. Above 50

```python
def determine_age_range(df):
    if df['age'] >= 51:
        val = 'Above 50'
    elif df['age'] >= 36 and df['age'] <= 50:
        val = '36-50'
    elif df['age'] >= 18 and df['age'] <= 35:
        val = '18-35'
    elif df['age'] <= 17:
        val = 'Under 18' 
    else:
        val = 'Unknown'
    return val
```

## Apply the function on the DataFrame and add a new column called "age_range"

```python
san_francisco_rides['age_range'] = san_francisco_rides.apply(determine_age_range, axis=1)
san_francisco_rides
```

# Complete multiple regression to see how trip duration is affected by age range, gender, and user type (using statsmodel)

```python
import statsmodels.formula.api as smf
```

```python
model = smf.ols(formula = 'trip_duration_sec ~ age_range + member_gender + user_type', data = san_francisco_rides)
results = model.fit()
print(results.summary())
```
